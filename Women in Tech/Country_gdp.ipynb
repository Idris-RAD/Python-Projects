```python
countrygdp = pd.read_csv('C:\\Users\\Arty\\Desktop\\julia\\IronHack\\Datasets\\stats\\countrygdp2.csv', sep=';')
```

```python
result = pd.merge(countrylist, countrygdp, how='left', on='country')
```
